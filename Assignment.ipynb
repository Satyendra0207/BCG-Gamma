```python
def company_price(newdata):
  company=newdata.split(",")[0]
  #max=company.filter()
  fuel=newdata.split(",")[1]
  engine_type=newdata.split(",")[11]
  price=newdata.split(",")[19]

  return  company+"        " + price +"      " + fuel +"     " +engine_type
```

```python
data=newdf.map(company_price)
```

```python
data.take(4)
```
